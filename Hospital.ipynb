```python
class Patient:
    def __init__(self, name, age, gender, admitted=False):
        self.name = name
        self.age = age
        self.gender = gender
        self.admitted = admitted


class Doctor(Patient):
    def __init__(self, name, age, gender, specialty, admitted=False):
        super().__init__(name, age, gender, admitted)
        self.specialty = specialty

    def prescribe_medication(self, patient):
        if patient.admitted:
            print(f"Dr. {self.name} prescribes medication for {patient.name}.")
        else:
            print(f"{patient.name} is not admitted. Prescribing medication is not possible.")

class Hospital:
    def __init__(self):
        self.patients = []

    def admit_patient(self, name, age, gender):
        patient = Patient(name, age, gender, admitted=True)
        self.patients.append(patient)
        print(f'{patient.name} admitted to the hospital.')

    def discharge_patient(self, name):
        for patient in self.patients:
            if patient.name.lower() == name.lower() and patient.admitted:
                patient.admitted = False
                print(f'{patient.name} discharged from the hospital.')
                return
        print(f'Sorry, patient "{name}" not found or not currently admitted.')

    def display_patients(self):
        print("\nList of Patients in the Hospital:")
        for patient in self.patients:
            status = "Admitted" if patient.admitted else "Not Admitted"
            print(f'{patient.name} - Age: {patient.age}, Gender: {patient.gender} - {status}')


    def add_doctor(self, name, age, gender, specialty):
        doctor = Doctor(name, age, gender, specialty)
        self.doctors.append(doctor)
        print(f'Dr. {doctor.name} added to the hospital.')

    def display_doctors(self):
        print("\nList of Doctors in the Hospital:")
        for doctor in self.doctors:
            print(f'Dr. {doctor.name} - Age: {doctor.age}, Gender: {doctor.gender}, Specialty: {doctor.specialty}')

def main():
    hospital = Hospital()

    while True:
        print("\nHospital Management System")
        print("1. Admit Patient")
        print("2. Display Patients")
        print("3. Discharge Patient")
        print("4. Add Doctor")
        print("5. Display Doctors")
        print("6. Exit")

        choice = input("Enter your choice (1-6): ")

        if choice == "1":
            name = input("Enter the patient's name: ")
            age = input("Enter the patient's age: ")
            gender = input("Enter the patient's gender: ")
            hospital.admit_patient(name, age, gender)

        elif choice == "2":
            hospital.display_patients()

        elif choice == "3":
            name = input("Enter the name of the patient you want to discharge: ")
            hospital.discharge_patient(name)

        elif choice == "4":
            print("Exiting the Hospital Management System. Goodbye!")
            break
        elif choice == "5":
            hospital.display_doctors()

        elif choice == "6":
            print("Exiting the Hospital Management System. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 1 and 6.")

if __name__ == "__main__":
    main()


```
